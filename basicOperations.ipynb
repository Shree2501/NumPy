BASIC OPERATIONS

```python
import numpy as np
a1=np.array([1,3,4])
a2=np.array([5,6,3])
print(a1**2)
print(a1+a2)
print(a2-a1)
print(a1*a2)
print(a1/a2)
print(a1.dot(a2))
print(np.sqrt(a1))
print(np.std(a2))
```

DIMENTIONS

```python
import numpy as np
a=np.array([1,3,4])
b=np.array([[1,3],[2,4],[6,7]])
print(a)
print(np.ndim(a))
print('dimensions:',np.ndim(b))
```

DATA TYPE / SIZE / SHAPE

```python
import numpy as np
a=np.array([[1,3],[2,4],[6,7]])
print(a)
print('data type:',a.dtype)    #shows datatype
b=a.astype(float)      #typecasting
print(b)
print('data type after:',b.dtype)
print('size of element:',a.itemsize,"Bits")
print('size of array:',a.size)
print('shape:',a.shape)
print(a.reshape(2,3))
print(a.reshape(6,1))
```
