RAVEL / FLATEN FUNCTION

```python
import numpy as np
a=np.array([[1,3],[2,4],[6,7]])
print(a)
#view
print(a.ravel())
print(a.flatten())
```

NDITER FUNCTION

```python
import numpy as np
a=np.arange(12).reshape(3,4)
print(a)
print()
#iterate elements column vise
for x in np.nditer(a,order='F'):#F stands for fortran
  print(x,end=" ")
print()
#transpose column into raws and rows into columns
for x in np.nditer(a,order='F',flags=['external_loop']):
  print(x)
print()
for x in np.nditer(a,op_flags=['readwrite']):
  x[...]=x*x
print(a)
```
